```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("fish.csv")
df.head(5)
```

```python
df.columns
```

```python
df_reduced = df.drop(df.loc[:,'NH Jan':"SH Dec"].columns,axis=1)
df_reduced
```

```python
df_reduced.nunique()
```

```python
corr = df_reduced.corr()
sns.heatmap(corr, annot=True)
```

There's a high positive correlation between `Total Catches to Unlock` and `Sell`. We will use this for regression later on.

```python
import missingno
plt.figure(figsize=(10,10))
missingno.matrix(df_reduced, figsize=(10,4))
plt.show()
```

```python
sns.boxplot(x = df_reduced["Sell"])
```

```python
q1, q3= np.percentile(df_reduced["Sell"], 0.25), np.percentile(df_reduced["Sell"], 0.75)
IQR = q3 - q1
upper_bound = q3 + 1.5 * IQR
outliers = df_reduced[df_reduced["Sell"] >upper_bound]
print(upper_bound, q3, q1)
outliers
```
